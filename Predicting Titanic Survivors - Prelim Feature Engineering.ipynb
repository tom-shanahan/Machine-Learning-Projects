```python
# transforming test data

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

g_sub = pd.read_csv('gender_submission.csv')
test = pd.read_csv('test.csv')
train = pd.read_csv('train.csv')

test_df = pd.DataFrame(data = test)
train_df = pd.DataFrame(data = train)

pclass_dummies = pd.get_dummies(train_df['Pclass'])
embarked_dummies = pd.get_dummies(train_df['Embarked'])

train_df1 = pd.concat([pclass_dummies, embarked_dummies], axis = 1)
train_df1 = pd.concat([train_df, train_df1], axis = 1)

# Break up last name, first name, and title
name = train_df['Name'].tolist()
last_first = []
for i in range(len(name)):
    last_first.append(name[i].split(", ",1))

first = []
for i in range(len(last_first)):
    first.append(last_first[i][1])
    
last = []
for i in range(len(last_first)):
    last.append(last_first[i][0])
    
title = []
for i in range(len(first)):
    title.append((first[i].split(" ",1))[0])
    
title_df = pd.DataFrame(data = title)
last_df = pd.DataFrame(data = last)
train_df1['Name'] = title_df[0]
train_df1 = pd.concat([train_df1, last_df], axis = 1)

# created dummy variables for ticket classes and embarked location
train_df1.drop(['Embarked'], axis = 1, inplace = True)
train_df1.drop(['Pclass'], axis = 1, inplace = True)

# gapminder.rename(columns={'pop':'population'}, inplace=True)
train_df1.rename(columns = {1:'First_Class', 2:'Second_Class', 3:'Third_Class', 
                            'C':'C_Embarked', 'Q':'Q_Embarked', 'S':'S_Embarked',
                           'Name':'Title', 0:'Last Name'}, inplace = True)
train_df1.to_csv('better_train.csv', index = False)

# Dropped Ticket and Cabin 
train_df1.drop(['Cabin'], axis = 1, inplace = True)
train_df1.drop(['Ticket'], axis = 1, inplace = True)
train_df1['Title'] = train_df1['Title'].replace('Mlle.', 'Ms.')
train_df1['Title'] = train_df1['Title'].replace('Mme.', 'Mrs.')
train_df1['Title'] = train_df1['Title'].replace('Ms.', 'Miss.')
train_df1['Title'] = train_df1['Title'].replace(['Lady.', 'Countess.','Capt.', 'Col.', 'Don.', 'Dr.', 
                                                 'Major.', 'Rev.', 'Sir.', 'Jonkheer.', 'the', 'Dona.'], 'Rare')
# Created Family size from Parch and SibSp

train_df1['Family Size'] = train_df1['Parch'] + train_df1['SibSp'] + 1
train_df1.drop(['SibSp'], axis = 1, inplace = True)
train_df1.drop(['Parch'], axis = 1, inplace = True)

# Created new column to determine if the person was alone or not
train_df1['IsAlone'] = 0
train_df1.loc[train_df1['Family Size'] == 1, 'IsAlone'] = 1
train_df1.drop(['Family Size'], axis = 1, inplace = True)

        
```
